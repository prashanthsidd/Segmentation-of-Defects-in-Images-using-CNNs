```
#invert all the images in gt
def invert_GT_images(path='./dataset/CFgt'):
  for file in os.listdir(path):
    file_name = path + file
    img = cv.imread(file_name)
    img_inv = cv.bitwise_not(img)

    cv.imwrite(file_name, img_inv)
```

```
def data_aug(image_path, gt_path):
  
  imgs = os.listdir(image_path)
  counter = len(imgs)
  print('counter is {}'.format(counter))
  for img in imgs:
  
    gt_name = img[:-4] + '.png'
    
    im = cv.imread(os.path.join(image_path, img))  
    gt = cv.imread(os.path.join(gt_path, gt_name))  
    
    assert im is not None and gt is not None 
    
    #Flip
    im_flip = np.fliplr(im)
    gt_flip = np.fliplr(gt)
    
    cv.imwrite(os.path.join(image_path, str(counter) + img[-4:]), im_flip)
      
    cv.imwrite(os.path.join(gt_path, str(counter) + '.png'), gt_flip)
    
    counter += 1
    
    #Rotate 90 degrees 3 times
    for i in range(3):
      
      im = np.rot90(im)
      gt = np.rot90(gt)
      
      cv.imwrite(os.path.join(image_path, str(counter) + img[-4:]), im)
      
      cv.imwrite(os.path.join(gt_path, str(counter) + '.png'), gt)
      
      counter += 1
      
      im_flip = np.rot90(im_flip)
      gt_flip = np.rot90(gt_flip)
      
      cv.imwrite(os.path.join(image_path, str(counter) + img[-4:]), im_flip)
      
      cv.imwrite(os.path.join(gt_path, str(counter) + '.png'), gt_flip)
      
      counter += 1
      
      

      
```
